## Data Preparation

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img

import os
import pandas as pd
import random
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train = pd.read_csv("D:/Documents/CVPR/DataSets/Kannada-MNIST/train.csv")

train.tail()
```

```python
test = pd.read_csv("D:/Documents/CVPR/DataSets/Kannada-MNIST/Dig-MNIST.csv")

test.tail()
```

```python
x_train = train.iloc[:, 1:].to_numpy()
y_train = train.iloc[:, 0].to_numpy()

print(x_train.shape)
print(y_train.shape)
```

```python
x_test = test.iloc[:, 1:].to_numpy()
y_test = test.iloc[:, 0].to_numpy()

print(x_test.shape)
print(y_test.shape)
```

## Normalise Data

```python
x_train, x_test = x_train.astype("float32") / 255.0, x_test.astype("float32") / 255.0
x_train = x_train.reshape(-1, 28, 28, 1)
x_test = x_test.reshape(-1, 28, 28, 1)
print(x_train.shape)
print(x_test.shape)
```

## Validation Split

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=2)
print(x_train.shape)
print(x_val.shape)
```


## Sample Data

```python
i = 0
plt.figure(figsize=(20,3))
for i in range(60):
    plt.subplot(3, 20, i+1)
    plt.imshow(x_train[i], cmap=plt.cm.gray_r)
        
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(y_train[i])
plt.tight_layout()
plt.show()
```

## Data Augmentation

```python
datagen = ImageDataGenerator(
    rotation_range=12,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.2,
    zoom_range=0.3,
    horizontal_flip=False,
    vertical_flip=False,
    fill_mode='nearest'
)
```

## CNN Model

```python
model = keras.Sequential([
    keras.Input(shape=x_train.shape[1:]),
    layers.Conv2D(filters=32, kernel_size=(5,5),strides=(1,1), padding='valid', activation='relu'),
    layers.MaxPool2D(pool_size=(2,2),strides=(1,1), padding='same'),
    layers.Conv2D(filters=64, kernel_size=(3,3),strides=(1,1), padding='valid', activation='relu'),
    layers.MaxPool2D(pool_size=(2,2),strides=(1,1), padding='same'),
    layers.Flatten(),
    layers.Dense(units=128, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(units=10, activation='softmax')
])
model.summary()
```

```python
model.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)
```

## Train

```python
#cb = EarlyStopping(monitor='val_loss', min_delta=0.2, patience=5, restore_best_weights=True)

h = model.fit(datagen.flow(x_train, y_train, batch_size=64), 
              epochs=5, 
              validation_data=(x_val,y_val)
             )
```

## Ploting the Training result

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(h.history['accuracy'], 'o-', label='train accuracy')
plt.plot(h.history['val_accuracy'], 'o-', label = 'validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(h.history['loss'], 'o-', label='train loss')
plt.plot(h.history['val_loss'], 'o-', label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend(loc='upper right')

plt.show()
```

## Model Evaluation

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('\nTest accuracy:', test_acc)
```

## Re-Prepare Data

```python
train = pd.read_csv("D:/Documents/CVPR/DataSets/Kannada-MNIST/train.csv")
test = pd.read_csv("D:/Documents/CVPR/DataSets/Kannada-MNIST/Dig-MNIST.csv")
x_train = train.iloc[:, 1:].to_numpy()
y_train = train.iloc[:, 0].to_numpy()

print(x_train.shape)
print(y_train.shape)

x_test = test.iloc[:, 1:].to_numpy()
y_test = test.iloc[:, 0].to_numpy()

print(x_test.shape)
print(y_test.shape)
```

## Normalise Data

```python
x_train = x_train.reshape(-1, 28, 28,1)
x_test = x_test.reshape(-1, 28, 28,1)
x_train, x_test = x_train.astype("float32") / 255.0, x_test.astype("float32") / 255.0
print(x_train.shape)
print(x_test.shape)
```

## Validation Split

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=2)
print(x_train.shape)
print(x_val.shape)
```

## Data Augmentation

```python
train_datagen = ImageDataGenerator(
                                   rotation_range=10,
                                   width_shift_range=0.25,
                                   height_shift_range=0.25,
                                   shear_range=0.1,
                                   zoom_range=0.25,
                                   horizontal_flip=False)

valid_datagen = ImageDataGenerator(
                                    horizontal_flip=False,
                                    rotation_range=15,
                                   width_shift_range=0.25,
                                   height_shift_range=0.25,
                                   shear_range=0.15,
                                   zoom_range=0.25,
                                    )
```

## CNN Model 3

```python
model3 = keras.Sequential([
    keras.Input(shape=x_train.shape[1:]),
    layers.Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu'),
    #layers.BatchNormalization(),
    layers.Conv2D(filters=64, kernel_size=(5,5), padding='same', activation='relu'),
    #layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2)),
    
    layers.Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu'),
    #layers.BatchNormalization(),
    layers.Conv2D(filters=128, kernel_size=(5,5), padding='same', activation='relu'),
    #layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2)),
    layers.Dropout(0.15),
    
    layers.Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu'),
    #layers.BatchNormalization(),
    
    layers.GlobalMaxPooling2D(),
    layers.Dropout(0.15),
    
    layers.Flatten(),
    layers.Dense(units=512, activation='relu'),
    layers.Dense(units=128, activation='relu'),
    layers.Dense(units=10, activation='softmax')
])
model3.summary()
```

## Compling Model 3

```python
model3.compile(
    optimizer='rmsprop', 
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)
```

## Train Model 3

```python
h = model3.fit(x=x_train, y=y_train, epochs=6, batch_size=64, validation_split=0.3)
```

## Train Accuracy of Model 3

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(h.history['accuracy'], 'o-', label='train accuracy')
plt.plot(h.history['val_accuracy'], 'o-', label = 'validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(h.history['loss'], 'o-', label='train loss')
plt.plot(h.history['val_loss'], 'o-', label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend(loc='upper right')

plt.show()
```

## Model 3 Evaluate

```python
test_loss, test_acc = model3.evaluate(x_test, y_test)
print('\nTest accuracy:', test_acc)
```

## CNN Model 4

```python
model4 = keras.Sequential([
    keras.Input(shape=x_train.shape[1:]),
    layers.Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(filters=64, kernel_size=(3,3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(filters=64, kernel_size=(5,5), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2)),
    
    layers.Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(filters=128, kernel_size=(3,3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(filters=128, kernel_size=(5,5), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPool2D(pool_size=(2,2)),
    layers.Dropout(0.15),
    
    layers.Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    layers.Conv2D(filters=256, kernel_size=(3,3), padding='same', activation='relu'),
    layers.BatchNormalization(),
    
    layers.MaxPool2D(pool_size=(2,2)),
    layers.Dropout(0.15),
    
    layers.Flatten(),
    layers.Dense(units=512, activation='relu'),
    layers.Dense(units=128, activation='relu'),
    layers.Dense(units=10, activation='softmax')
])
model4.summary()
```

## Model 4 Compile

```python
model4.compile(
    optimizer='adam', 
    loss='sparse_categorical_crossentropy', 
    metrics=['accuracy']
)
```

## Model 4 Train

```python
cb = EarlyStopping(monitor='val_loss', min_delta=0.2, patience=5, restore_best_weights=True)

h = model4.fit(train_datagen.flow(x_train, y_train, batch_size=64), 
              epochs=20, 
              validation_data=valid_datagen.flow(x_val,y_val),
              callbacks=[cb]
             )
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plt.plot(h.history['accuracy'], 'o-', label='train accuracy')
plt.plot(h.history['val_accuracy'], 'o-', label = 'validation accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.grid(True)
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(h.history['loss'], 'o-', label='train loss')
plt.plot(h.history['val_loss'], 'o-', label='validation loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.grid(True)
plt.legend(loc='upper right')

plt.show()
```

```python
test_loss, test_acc = model4.evaluate(x_test, y_test)
print('\nTest accuracy:', test_acc)
```
